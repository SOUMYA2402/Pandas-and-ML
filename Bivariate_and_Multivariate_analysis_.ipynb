# **1. Scatterplot (Numerical-Numerical)**

- **here each dot is representing each customer, such that the relation btw total bill n tips is linear, more the total bill, more is the tips**

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
```

- **Now this is a multivariate analysis coz we r using 3 columns, tips, total_bill, sex**

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex',style='smoker')
```

- **Again a multivariate analysis but this time with 4 columns, sex, smoker, total_bill and tips**

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='size')
```

**- size col represents no. of members with that person, again the multivariate with 5 variables**

## **5. HeatMap (Categorical - Categorical)**

```python
titanic.head()
```

**pandas.crosstab() is a function in Python's pandas library that creates a cross-tabulation table (also known as a contingency table). Its used to show the frequency (count) of occurrences for combinations of two (or more) categorical variables.**

```python
df=pd.crosstab(titanic['Pclass'],titanic['Survived'])
df.head()
```

```python
sns.heatmap(df)
```

alternaive syntax for it is:

```python
# sns.heatmap(index='Pclass', column='Survived',data=titanic)
# We can't use this syntax, we need to build cross table
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100)
```

```python
(titanic.groupby('Embarked')['Survived'].mean()*100)
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```
