## **Ordinal Encoding for Ordinal Categorical Data**

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("customer.csv")
```

```python
df.head()
```

###**Encoding is used to convert ordinal categorical data to numerical data**

- review and education are ordinal data, where max no. has to give max number.

- gender column mae one hot encoding lagana hai and purchased column output column hai uspar label encoding lagegi

- age column par imputer lagega

- -saara cols par alag alag lagega isleye humlog review and education ko side par lange and uss par ordinal encoding karenge


```python
df.columns
```

```python
df=df[['review','education', 'purchased']]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
X_train.head()
```

```python
X_train['education'].value_counts()
```

```python
X_train['education'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Poor', 'Average', 'Good'],['School', 'UG', 'PG']])
# the order id necessary , poor gets least no n good gets max no., similarly in school and pg
oe.fit(X_train)
X_train=oe.transform(X_train)
X_test=oe.transform(X_test)
# Also, once you've transformed X_train, it no longer has strings but
# just numbers; so running the same .transform(X_train) again would fail because it expects strings.

```
